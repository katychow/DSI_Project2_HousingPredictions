```python
numerical_train = train._get_numeric_data()
```

```python
X = numerical_train.drop(columns=['Id','PID','SalePrice']).copy()
X.fillna(0, inplace = True)

y=numerical_train['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
pipe_lasso = Pipeline([
    ('pf',PolynomialFeatures()),
    ('ss',StandardScaler()),
    ('lasso',LassoCV(max_iter=10000))
])
```

```python
pipe_lasso.fit(X_train,y_train)
```

```python
# plt.plot(np.unique(train['SalePrice']), np.poly1d(np.polyfit(train['SalePrice'], train['Gr Liv Area'], 1))(np.unique(train['SalePrice'])))
```

```python
plt.figure(figsize = (8,6))

plt.scatter(y_test, pipe_lasso.predict(X_test), color ='pink', alpha = 0.3, marker = 'H',s=100)
plt.xlabel('True Home Value', size = 12)
plt.ylabel('Predicted Home Value', size = 12)
plt.title('True vs Predicted Home Value in Ames, Iowa\nModel 2', size = 16)
plt.plot(np.unique(y_test), np.poly1d(np.polyfit(y_test, pipe_lasso.predict(X_test), 1))(np.unique(y_test)),'--',color = 'gray',alpha=0.5)

plt.savefig('Model2_true_v_pred_prices.png')
```

```python
pipe_lasso.score(X_train,y_train), pipe_lasso.score(X_test,y_test)
```

```python
pd.DataFrame({
    'coefs': pipe_lasso.named_steps['pf'].get_feature_names(X_train.columns),
    'value': pipe_lasso.named_steps['lasso'].coef_
}).set_index('coefs').sort_values('value', ascending=False).to_csv('/Users/katychow/Documents/katy_sub4_lasso_coefs.csv')
```
