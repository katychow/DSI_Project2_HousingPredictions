```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error

%matplotlib inline
```

```python
train = pd.read_csv('/Users/katychow/DSI/Projects/project-2/datasets/train.csv',na_values='',keep_default_na=False)
test = pd.read_csv('/Users/katychow/DSI/Projects/project-2/datasets/test.csv',na_values='',keep_default_na=False)

train = train[(train['Gr Liv Area']<4000) & (train['Lot Area']<25000)]
```

```python
numerical_train = train._get_numeric_data()
X = numerical_train.drop(columns=['Id','PID','SalePrice']).copy()
X.fillna(0, inplace = True)

y=numerical_train['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
pipe_ridge = Pipeline([
#     ('pf',PolynomialFeatures()),
    ('ss',StandardScaler()),
    ('ridge',RidgeCV())
])

pipe_lasso = Pipeline([
#     ('pf',PolynomialFeatures()),
    ('ss',StandardScaler()),
    ('lasso',LassoCV(max_iter=10000))
])
```

```python
pipe_lasso.fit(X_train,y_train)
pipe_ridge.fit(X_train,y_train)
```

```python
pipe_lasso.score(X_train,y_train), pipe_lasso.score(X_test,y_test)
```

```python
pipe_ridge.score(X_train,y_train), pipe_ridge.score(X_test,y_test)
```

```python
np.sqrt(mean_squared_error(y, pipe_lasso.predict(X)))
```
